```python
from tensorflow.keras.layers import Embedding,LSTM,Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import one_hot
```

```python
## Vocabulary size
voc_size=5000
```

## One Hot Represenatation

```python
message=X.copy()
```

```python
message['title']
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
message.reset_index(drop=True, inplace=True)
```

```python
corpus=[]

for i in range(0,len(message)):
    review=re.sub('[^a-zA-Z]',' ',message['title'][i])
    review=review.lower()
    review=review.split()
    
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

## One Hot Representation

```python
onehot_repr=[one_hot(words,voc_size) for words in corpus]
```

```python
onehot_repr
```

```python
corpus[1]
```

```python
onehot_repr[1]
```

## Embedding Representation

```python
sent_length=20
```

```python
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
```

```python
embedded_docs
```

## Creating the LSTM Model

```python
embedding_vector_features=40 #Embedded docs is converted into 40 vectors
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
len(embedded_docs),y.shape
```

```python
import numpy as np
```

```python
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_final,y_final,test_size=0.2)
```

```python
## Finally Training the Model

model.fit(X_final,y_final,validation_data=(X_test,y_test),epochs=20,batch_size=64)
```

## Adding Dropout
